## 2. Diurnal temperature change

A simple model of diurnal surface temperature variation is
$$ \frac{dT}{dt} = S(t)-\gamma(T-T_{eq})$$
where $S(t)$ is the rate of solar heating, and $\gamma$ is the rate at which temperature $T$ relaxes towards its equilibrium temperature $T_{eq}$. The second term on the right-hand side represents longwave radiative cooling. 

Solar heating is roughly sinusoidal
$$ S(t) = S_0 \max\left(0,\cos\left(\frac{2\pi t}{24}\right)\right) $$
where $S_0$ is the peak heating rate at local noon. The max function ensures that solar heating is always zero or greater.

For this problem, use constants
* $S_0 = 2 {\rm K h^{-1}}$
* $\gamma = 0.1 {\rm h^{-1}}$
* $T_{eq} = 300 {\rm K}$
