## 3. Vertical layers of the atmosphere

The hypsometric equation specifies that the thickness ($Z_2-Z_1$) of an air layer extending from pressure $p_1$ at the bottom to $p_2$ at the top is
$$ Z_2-Z_1 = \frac{R_d \bar{T_v}}{g}\ln\left(\frac{p_2}{p_1}\right)$$
where $\bar{T_v}$ is the mean virtual temperature of the layer, $R_d=287 {\rm J\,kg^{-1}\, K^{-1}}$ is the specific gas constant, and $g=9.81 {\rm m\,s^{-2}}$ is gravitational acceleration.
