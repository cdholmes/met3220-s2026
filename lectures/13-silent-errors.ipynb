## Sea surface temperature

What is the global mean sea-surface temperature? 
Use OISST data

```python
import xarray as xr
url = 'http://psl.noaa.gov/thredds/dodsC/Datasets/noaa.oisst.v2/sst.mnmean.nc'
oisst = xr.open_dataset(url)

oisst
```

```python
# Time of interest
t = '1997-12-01'

print( f'Global mean SST for {t}:',
       oisst['sst'].sel(time=t)\
            .mean(dim=['lon','lat']).values,
       oisst['sst'].attrs['units'])
```

Code ran without errors. Resulting temperature seems plausible.

## Annual Precipitation for 2005

Three versions are below.

```python
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Read in the NARR data
file = '../assignments/prog11_files/narr.1995-2025.apcp.mon.mean.nc'
narr_monthly = xr.open_dataset(file)

# Make all longitudes negative for plotting
narr_monthly['lon'] = narr_monthly['lon'] % 360 - 360
# kg/m2 is equivalent to mm
narr_monthly['apcp'].attrs['units'] = 'mm'

# Compute annual precipitation
narr_annual = narr_monthly.resample(time='1YS').mean()
narr_annual['apcp'].attrs['long_name'] = 'Annual Precipitation'
narr_annual['apcp'].attrs['units'] = 'mm'

# Plot the 2005 annual precipitation
p = narr_annual['apcp']\
        .sel(time='2005')\
        .plot(x='lon', y='lat', 
            vmax=15, vmin=0,
            transform=ccrs.PlateCarree(),
            subplot_kws={'projection':ccrs.LambertConformal(central_latitude=39, 
                                            central_longitude=-100)}, 
            cmap='YlGnBu')
p.axes.set_extent([-125, -65, 25, 70], crs=ccrs.PlateCarree())
p.axes.coastlines()
p.axes.add_feature(cfeature.BORDERS)
p.axes.set_title('NARR Annual Precipitation, 2005') 
```

```python
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Read in the NARR data
file = '../assignments/prog11_files/narr.1995-2025.apcp.mon.mean.nc'
narr_monthly = xr.open_dataset(file)

# Make all longitudes negative for plotting
narr_monthly['lon'] = narr_monthly['lon'] % 360 - 360
# kg/m2 is equivalent to mm
narr_monthly['apcp'].attrs['units'] = 'mm'

# Compute annual precipitation
narr_annual = (narr_monthly * narr_monthly['time'].dt.days_in_month)\
                .resample(time='1YS').sum()
narr_annual['apcp'].attrs['long_name'] = 'Annual Precipitation'
narr_annual['apcp'].attrs['units'] = 'mm'

# Plot the 2005 annual precipitation
p = narr_annual['apcp']\
        .sel(time='2005')\
        .plot(x='lon', y='lat', 
            vmax=3000, vmin=0,
            transform=ccrs.PlateCarree(),
            subplot_kws={'projection':ccrs.LambertConformal(central_latitude=39, 
                                            central_longitude=-100)}, 
            cmap='YlGnBu')
p.axes.set_extent([-125, -65, 25, 70], crs=ccrs.PlateCarree())
p.axes.coastlines()
p.axes.add_feature(cfeature.BORDERS)
p.axes.set_title('NARR Annual Precipitation, 2005') 
```
