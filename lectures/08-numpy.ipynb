## Broadcasting


Not all the arrays we want to work with will have the same size.
One approach would be to manually "expand" our arrays to all be the same size, e.g. using `tile`.
_Broadcasting_ is a more efficient way to multiply arrays of different sizes
Numpy has specific rules for how broadcasting works.
These can be confusing but are worth learning if you plan to work with Numpy data a lot.

The core concept of broadcasting is telling Numpy which dimensions are supposed to line up with each other.

Example 1: Array + Scalar

<img src="https://numpy.org/doc/stable/_images/broadcasting_1.png" style="background-color:white;">

Example 2: Array + Array

<img src="https://numpy.org/doc/stable/_images/broadcasting_2.png" style="background-color:white;">

<img src="https://numpy.org/doc/stable/_images/broadcasting_4.png" style="background-color:white;">

Broken example:

<img src="https://numpy.org/doc/stable/_images/broadcasting_3.png" style="background-color:white;">

[Numpy broadcasting documentation](https://numpy.org/doc/stable/user/basics.broadcasting.html)

Dimensions are automatically aligned _starting with the rightmost (last) dimension_.
If dimensions have the same length or one of them is length 1, then the two arrays can be broadcast.

````python
# Create arrays similar to the example
a = np.array( [0, 10, 20, 30] )
b = np.array( [1, 2, 3] )
print(a.shape, b.shape)

# This broadcast attempt won't work because the rightmost dimensions don't align
c = a + b
````

We can add a dimension to enable broadcasting

````python
# We can add a dimension to enable broadcasting
a_column = a[:,np.newaxis]
print(a_column.shape, b.shape)
print(a_column + b)
````

````python
# Meteorology example
# Barometric equation: pressure decreases exponentially with altitude
# p(z) = p0 * exp(-z/H), where H is the scale height (about 7500 m)

# We have surface pressure measurements at 4 times, hPa
p_surface = np.array( [1010, 1012, 1015, 1013] )
# We want to compute the pressure at 3 altitudes: 500, 1000, 1500 m for each time
altitude = np.array( [500, 1000, 1500] )

print(p_surface.shape, altitude.shape)
````

````python
# Make altitude a column vector to enable broadcasting
altitude_column = altitude[:, np.newaxis]
print(altitude_column.shape)
````

````python
# Apply the barometric equation to compute pressure at every altitude and every time
# result is a 3x4 array: 3 altitudes, 4 times
H = 7500
pressure_at_altitude = p_surface * np.exp( -altitude_column / H )
print(pressure_at_altitude)
````

### Example

Let's revisit the calculation of an accelerating car (iteration notebook). The position change of the car is 
$$\frac{dx}{dt} = v.$$
In finite difference form
$$\frac{\Delta x}{\Delta t} = v \;{\rm or}\; \Delta x = v \Delta t $$
Given a position $x_i$ at time $t_i$, the new position after an interval $\Delta t$ will be
$$x_{i+1}=x_i + \Delta x = x_i + v\Delta t.$$
In this specific example, we assume that the car starts at $x=0$ at $t=0$ and its velocity is $v=2t$.

````python
# Simulate the motion of a car with a velocity that increases linearly with time

# Number of time steps
nsteps = 500
# Time step size, s
dt = 0.01

# Arrays to hold time and position of the car
time = np.arange(nsteps+1) * dt
x = np.zeros(nsteps+1)

# Initial position of the car at t=0
x[0] = 0

for i in range(nsteps):
    # velocity at time step i
    v = 2 * time[i]  
    # Update the position of the car at time step i
    x[i+1] = x[i] + v * dt
    
    # We no longer need to set x_old = x_new every time step
    # because we are storing the position at every time step in the array x

# Plot the position as a function of time
plt.plot(time, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
````

## Derivatives

Computing derivatives is a common task in scientific analyses. For example, the pressure gradient ($\nabla p$) in the pressure-gradient force is a derivative in two directions. The diffusion, advection, and wave equations also involve derivatives. We will examine how to compute derivatives from discrete data (points, not continuous functions).

Suppose we have a 1-D array `x` and corresponding values of a function `f` ($f(x)$) with points `i = 0...n`. Each element `f[i]` occurs at position `x[i]`. The derivative of `f` in finite difference form is
$$\frac{df}{dx}=\frac{\Delta f}{\Delta x}.$$

There are a few ways that we can compute the derivative at the point `x[i], f[i]`. 
* Forward finite difference $$\frac{\Delta f}{\Delta x} = \frac{f[i+1]-f[i]}{x[i+1]-x[i]} = \frac{f[i+1]-f[i]}{h}$$
* Backward finite difference $$\frac{\Delta f}{\Delta x} = \frac{f[i]-f[i-1]}{x[i]-x[i-1]} = \frac{f[i]-f[i-1]}{h}$$
* Centered finite difference $$\frac{\Delta f}{\Delta x} = \frac{f[i+1]-f[i-1]}{x[i+1]-x[i-1]} = \frac{f[i+1]-f[i-1]}{2h}$$
If the `x` points are evenly spaced, then $\Delta x$ is a constant, which we call $h$ in the equations above: `x[i+1]-x[i] = x[i]-x[i-1]= h`. 

The centered finite difference formula is generally the most accurate, but it only works for points `i = 1...n-1`. For the first point (`i=0`), only the forward finite difference formula can be used (because `i=-1` is not defined) and for the last point (`i=n`), only the backward finite difference formula can be used (because `i=n+1` is not defined).

````python
# Create some data
n = 100
x = np.linspace(0, 10, n)
f = x * np.sin(x)

# Compute the derivative df/dx
df_dx = np.zeros_like(f)

for i in range(n):
    if i == 0:
        # Forward difference for the first point
        df_dx[i] = (f[i+1] - f[i]) / (x[i+1] - x[i])
    elif i == n-1:
        # Backward difference for the last point
        df_dx[i] = (f[i] - f[i-1]) / (x[i] - x[i-1])
    else:
        # Central difference for the interior points
        df_dx[i] = (f[i+1] - f[i-1]) / (x[i+1] - x[i-1])

plt.plot(x,f)
plt.plot(x, df_dx)
````

When $f(x,y)$ is 2D, we can compute derivatives in each direction: $\frac{\partial f}{\partial x}$, $\frac{\partial f}{\partial y}$. 

````python
# Generate some data for x (longitude), y (latitude)
x = np.linspace(-180, 180, 200)
y = np.linspace(-90,  90,  100)
y_column = y[:, np.newaxis]

f2D = ( np.sin( x * np.pi/180 ) * np.cos( 0.5 * y_column * np.pi/180) )**2
````

````python
# Derivative df/dx for a function
ny, nx = f2D.shape  # number of rows, number of columns

# Grid is uniform, so we can use a constant grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Initialize an array to hold the derivative values
# df2D_dx = np.zeros_like(f2D)
df2D_dx = np.zeros( (ny, nx) )

for j in range(ny):
    for i in range(nx):
        if i == 0:
            # Forward difference for the first column
            df2D_dx[j,i] = (f2D[j,i+1] - f2D[j,i]) / dx
        elif i == nx-1:
            # Backward difference for the last column
            df2D_dx[j,i] = (f2D[j,i] - f2D[j,i-1]) / dx
        else:
            # Central difference for the interior columns
            df2D_dx[j,i] = (f2D[j,i+1] - f2D[j,i-1]) / (2*dx)

plt.pcolormesh(x, y, df2D_dx)
plt.colorbar()
````

````python
# Numpy has built-in functions to compute derivatives using finite differences, 
# which are much faster than the explicit loops above
df2D_dx_fast = np.gradient(f2D, dx, axis=1)

plt.pcolormesh(x, y, df2D_dx_fast)
plt.colorbar()
````
